# População e amostra

<ul>
População: todo o conjunto
</ul>
<ul>
Amostra: um subconjunto do conjunto
    <ul>
    <li>sempre menos que o conjunto</li>
    <li>Mais rápido para processar</li>
    <li>Manor tempo para analisar</li>
    <li>Os numeros que são obtidos da amostra são estátisticas</li>
    <li>As amostras precisam ser <strong>randômica e representativa</strong></li>
    </ul>    
</ul>

## 3. amostragem sistemática


```python
len(df)
```

```python
n_amostras = len(df)//100
# 100 amostras
n_amostras
```

```python
random.seed(1)
# para fixar o numero aleatório
```

```python
amostra_sistematica = np.arange(random.randint(0,n_amostras),len(df),n_amostras)
amostra_sistematica
```

```python
def amostragem_sistematica(dataset,amostras):
    step = len(dataset)//amostras
    index = np.arange(random.randint(0,step),len(dataset),step)
    return dataset.iloc[index]
```

```python
df_sistematico = amostragem_sistematica(df,100)
df_sistematico
```

## 4. Amostragem por grupos
divide a população em grupos e seleciona um grupo especifico

```python
len(df)//10
#numero de grupos
```

```python
grupos = []
n_grupo = 0
cont =0
for _ in range(len(df)):
    grupos.append(n_grupo)
    cont +=1
    if cont > len(df)//10:
        cont=0
        n_grupo += 1    
```

```python
np.unique(grupos,return_counts=True)
```

```python
def amostragem_grupo(dataset,n_amostras):
    grupos = []
    id_grupo = 0
    cont =0
    n_grupos = len(dataset)//n_amostras
    for _ in range(len(dataset)):
        grupos.append(id_grupo)
        cont +=1
        if cont >= len(dataset)//n_grupos:
            cont=0
            id_grupo += 1
    dataset['grupos'] = grupos
    return dataset[dataset['grupos']== random.randint(0,n_grupos)]
```

```python
df_amostra_grupos = amostragem_grupo(df,100)
df_amostra_grupos
```

## 5. Amostragem estratificada

```python
# iremos utilizar uma biblioteca que possui uma função que nso returna uma divisão estratificada
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
df['income'].value_counts()
```

```python
income_split_porcentage = df['income'].value_counts()/len(df)
```

```python
income_split_porcentage
```

```python
split = StratifiedShuffleSplit(test_size=0.1)
for x, y in split.split(df,df['income']):
    df_x = df.iloc[x]
    df_y = df.iloc[y]
```

```python
df_y['income'].value_counts()/len(df_y)
```

```python
def amostragem_estratificada(dataset,n_amostras,coluna):
    split = StratifiedShuffleSplit(test_size=n_amostras/len(dataset))
    for x, y in split.split(dataset,dataset[coluna]):
        dataset_x = dataset.iloc[x]
        dataset_y = dataset.iloc[y]
    return dataset_y
```

```python
df_amostra_estratificada = amostragem_estratificada(df,100,'income')
df_amostra_estratificada
```

## 6.Amostra de reservatorio
<strong>Data stream</strong> de items com tamanho desconhecido que pode ser acessado somente uma vez, por exemplo uma base de dados de um e-commerce que é dinámico está aumantando a todo momento e não é constante

Algoritmo para sortear um item do stream, porém, cada item deve possuir a mesma probablidade de seleção

```python
def amostragem_reservatorio(dataset,n_amostras):
    stream = []
    tam = len(dataset)
    for i in range(tam):
        stream.append(i)
    
    i = 0
    reservatorio = [0] * n_amostras
    for i in range(n_amostras):
        reservatorio[i] = stream[i]
    
    while i < tam:
        j = random.randrange(i+1)
        if j < n_amostras:
            reservatorio[j]= stream[i]
        i +=1
    return dataset.iloc[reservatorio]    
```

```python
df_amostra_reservatorio = amostragem_reservatorio(df,100)
df_amostra_reservatorio
```
